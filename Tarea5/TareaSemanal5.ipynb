### **Red tipo T**

<img src="./RedT.PNG"  align="center" width="350"/>

#### **Parámetros impedancias Z**

Para la red topología T conectamos una fuente de corriente en la salida y otra en la entrada, de forma tal que la corriente "entre" al nodo común de las impedancias.

Ahora utilizamos la ley de mallas de Kirchoff para obtener las ecuaciones del circuito:

$$
V_1=I_1(Z_a+Z_b) + I_2Z_b
$$

$$
V_1=I_1Z_b + I_2(Z_c+Z_b)
$$

Calculamos las impedancias de la matriz Z:

$$
Z_{11}=\frac{V_1}{I_1}|_{I_2=0}=Z_a+Z_b      
\qquad
Z_{12}=\frac{V_1}{I_2}|_{I_1=0}=Z_b      
$$

$$
Z_{22}=\frac{V_2}{I_2}|_{I_1=0}=Z_c+Z_b
\qquad
Z_{21}=\frac{V_2}{I_1}|_{I_2=0}=Z_b
$$

Entonces la **matriz Z** queda:

$$
Z=
\begin{bmatrix}
Z_a+Z_b   &    Z_b  \\
Z_b   &   Z_c+Z_b     
\end{bmatrix}
$$

#### **Parámetros admitancias Y**

Para la matriz Y podemos transformar la matriz Z utilizando la siguiente expresión:

$$
Y=\frac{1}{det(Z)}* Adj(A)
$$

Siendo el det(A):
$$
det(Z)= (Z_a+Z_b)*(Z_c+Z_b)- Z_b^2
$$

$$
det(Z)= Z_aZ_c+Z_aZ_b+Z_bZ_c+Z_b^2- Z_b^2
$$

$$
det(Z)= Z_aZ_c+Z_aZ_b+Z_bZ_c
$$

y la Adj(Z):

$$
Adj(A)=
\begin{bmatrix}
Z_c+Z_b   &    -Z_b  \\
-Z_b   &   Z_a+Z_b     
\end{bmatrix}
$$

Reemplazando:

$$
Y=\frac{1}{ Z_aZ_c+Z_aZ_b+Z_bZ_c}*\begin{bmatrix} Z_c+Z_b   &    -Z_b  \\ -Z_b   &   Z_a+Z_b     \end{bmatrix}
$$

Entonces la **matriz Y** queda:

$$
Y=
\begin{bmatrix}
\frac{Z_c+Z_b}{ Z_aZ_c+Z_aZ_b+Z_bZ_c} & \frac{-Z_b}{ Z_aZ_c+Z_aZ_b+Z_bZ_c}   \\ \frac{-Z_b}{ Z_aZ_c+Z_aZ_b+Z_bZ_c}   & \frac{Z_a+Z_b}{ Z_aZ_c+Z_aZ_b+Z_bZ_c} 
\end{bmatrix}
$$

#### **Parámetros T**

Para transformar la matriz Z a la matriz T utilizamos la siguiente expresión:

$$
T=\frac{1}{Z_{21}} \begin{bmatrix} Z_{11} & det(Z) \\ 1 & Z_{22} \end{bmatrix}
$$

$$
T=\frac{1}{Z_b} \begin{bmatrix} Z_a+Z_b &  Z_aZ_c+Z_aZ_b+Z_bZ_c \\ 1 & Z_c+Z_b \end{bmatrix}
$$

Entonces la **matriz T** es:

$$
T=
\begin{bmatrix}
\frac{Z_a}{Z_b}+1 &  \frac{Z_aZ_c}{Z_b}+Z_a+Z_c \\ \frac{1}{Z_b} & \frac{Z_c}{Z_b}+1 
\end{bmatrix}
$$

### **Red tipo $\pi$**

<img src="./RedPi.PNG"  align="center" width="350"/>

#### **Parámetros admitancias Y**

Para esta red conectamos dos fuentes de tensión de la misma forma que en la red T, una en la entrada y otra en la salida.

Esta vez aplicamos la ley de nodos de Kirchoff:

$$
I_1=V_1(Y_A+Y_C)-V_2Y_C
$$

$$
I_2=V_2(Y_B+Y_C)-V_1Y_C
$$

Calculamos las admitancias de la matriz Y:

$$
Y_{11}=\frac{I_1}{V_1}|_{V_2=0}=Y_A+Y_C      
\qquad
Y_{12}=\frac{I_1}{V_2}|_{V_1=0}=-Y_C      
$$

$$
Y_{22}=\frac{I_2}{V_2}|_{V_1=0}=Y_B+Y_C
\qquad
Y_{21}=\frac{I_2}{V_1}|_{V_2=0}=-Y_C
$$

Entonces la **matriz Y** queda:

$$
Y=
\begin{bmatrix}
Y_A+Y_C    &    -Y_C \\ -Y_C   &  Y_B+Y_C      
\end{bmatrix}
$$

#### **Parámetros impedancias Z**

Para la matriz Z transformamos la matriz Z con la siguiente expresión:

$$
Z=\frac{1}{det(Y)}* Adj(Y)
$$

Siendo el det(Y):
$$
det(Y)= (Y_A+Y_C)*(Y_B+Y_C ) - Y_C^2
$$

$$
det(Y)= Y_AY_B+Y_AY_C+Y_CY_B+Y_C^2- Y_C^2
$$

$$
det(Y)= Y_AY_B+Y_AY_C+Y_CY_B
$$

y la Adj(Y):

$$
Adj(Y)=
\begin{bmatrix}
Y_B+Y_C    &    Y_C \\ Y_C   &  Y_A+Y_C      
\end{bmatrix}
$$

Reemplazando:

$$
Z=\frac{1}{Y_AY_B+Y_AY_C+Y_CY_B}*\begin{bmatrix} Y_B+Y_C    &    Y_C \\ Y_C   &  Y_A+Y_C \end{bmatrix}
$$

Entonces la **matriz Z** queda:

$$
Z=
\begin{bmatrix} \frac{Y_B+Y_C}{Y_AY_B+Y_AY_C+Y_CY_B} & \frac{Y_C}{Y_AY_B+Y_AY_C+Y_CY_B} \\ \frac{Y_C}{Y_AY_B+Y_AY_C+Y_CY_B} & \frac{Y_A+Y_C}{Y_AY_B+Y_AY_C+Y_CY_B} \end{bmatrix}
$$

#### **Parámetros T**

Para transformar la matriz Y a la matriz T utilizamos la siguiente expresión:

$$
T=-\frac{1}{Y_{21}} \begin{bmatrix} Y_{22} & 1 \\ det(Y) & Y_{11} \end{bmatrix}
$$

$$
T=-\frac{1}{-Y_C} \begin{bmatrix} Y_B+Y_C & 1 \\ Y_AY_B+Y_AY_C+Y_CY_B & Y_A+Y_C\end{bmatrix}
$$

Entonces la **matriz T** es:

$$
T=
\begin{bmatrix}
\frac{Y_B}{Y_C}+1 &  \frac{1}{Y_C} \\ \frac{Y_AY_B}{Y_C}+Y_A+Y_B & \frac{Y_A}{Y_C}+1 
\end{bmatrix}
$$

### **Red tipo Lattice**

#### **Parámetros impedancias Z**

Para hallar la matriz Z, cortocircuitamos los puertos necesarios para cada parámetro. Se nota a simpe vista que para la diagonal principal queda el paralelo entre la suma de las impedancias en serie.

$$
Z_{11}=Z_{22}= Z_A+Z_B // Z_A+Z_B
$$

Para $Z_{12}$ vemos que la corriente de las mallas son iguales al ser las impedancias de las mallas tambien iguales, por lo que queda definida por $I_2$ siendo $\frac{I_2}{2}$ la corriente de cada malla.

$$
V_1 + \frac{I_2}{2} Z_A = \frac{I_2}{2} Z_B
$$

$$
\frac{V_1}{I_2} = \frac{Z_B-Z_A}{2}
$$

Para $Z_{21}$ pasa exactamente lo mismo:

$$
V_2 + \frac{I_1}{2} Z_A = \frac{I_1}{2} Z_B
$$

$$
\frac{V_2}{I_1} = \frac{Z_B-Z_A}{2}
$$

Entonces los parámetros de la matriz son:

$$
Z_{11}=\frac{V_1}{I_1}|_{I_2=0}=\frac{Z_A+Z_B}{2}   
\qquad
Z_{12}=\frac{V_1}{I_2}|_{I_1=0}=\frac{Z_B-Z_A}{2}      
$$

$$
Z_{22}=\frac{V_2}{I_2}|_{I_1=0}=\frac{Z_A+Z_B}{2}
\qquad
Z_{21}=\frac{V_2}{I_1}|_{I_2=0}=\frac{Z_B-Z_A}{2}
$$

Entonces la **matriz Z** queda:

$$
Z=
\begin{bmatrix}
\frac{Z_A+Z_B}{2}    &   \frac{Z_B-Z_A}{2}  \\
\frac{Z_B-Z_A}{2}   &   \frac{Z_A+Z_B}{2}      
\end{bmatrix}
$$

#### **Parámetros admitancias Y**

Para la matriz Y simplemente cambiamos las impedancias por admitancias:

$$
Y_{11}=\frac{I_1}{V_1}|_{V_2=0}=\frac{Y_A+Y_B}{2}   
\qquad
Y_{12}=\frac{I_1}{V_2}|_{V_1=0}=\frac{Y_B-Y_A}{2}      
$$

$$
Y_{22}=\frac{I_2}{V_2}|_{V_1=0}=\frac{Y_A+Y_B}{2}
\qquad
Y_{21}=\frac{I_2}{V_1}|_{V_2=0}=\frac{Y_B-Y_A}{2}
$$

Entonces la **matriz Y** queda:

$$
Y=
\begin{bmatrix}
\frac{Y_A+Y_B}{2}     &   \frac{Y_B-Y_A}{2} \\
\frac{Y_B-Y_A}{2}  &   \frac{Y_A+Y_B}{2}       
\end{bmatrix}
$$

#### **Parámetros T**

Para estos parámetros aplicamos la conversión:

$$
T=\frac{1}{Z_{21}} \begin{bmatrix} Z_{11} & det(Z) \\ 1 & Z_{22} \end{bmatrix}
$$

Siendo el det(Z):

$$
det(Z)= (\frac{Z_A+Z_B}{2})^2 - (\frac{Z_B-Z_A}{2})^2
$$

$$
det(Z)= \frac{Z_A^2+2Z_AZ_B+Z_B^2}{4} - \frac{Z_B^2-2Z_AZ_B+Z_A^2}{4}
$$


$$
det(Z)= \frac{Z_A^2+2Z_AZ_B+Z_B^2 -Z_B^2+2Z_AZ_B-Z_A^2}{4}
$$

$$
det(Z)= Z_AZ_B
$$

Entonces:

$$
T=\frac{2}{Z_B-Z_A} \begin{bmatrix} \frac{Z_A+Z_B}{2}  &  Z_AZ_B \\ 1 & \frac{Z_A+Z_B}{2} \end{bmatrix}
$$

Entonces la **matriz T** es:

$$
T=
\begin{bmatrix}
\frac{Z_A+Z_B}{Z_B-Z_A} &  2\frac{Z_AZ_B}{Z_B-Z_A} \\ \frac{2}{Z_B-Z_A} & \frac{Z_A+Z_B}{Z_B-Z_A}
\end{bmatrix}
$$

### **Parámetros Z de la red con OPAMP**

Del circuito podemos deducir las siguientes expresiones:

$$
I_1=\frac{V_1}{R1}
$$

$$
\frac{V_1}{R1}=-\frac{V_2}{R2}
$$

Las cuales nos permiten calcular los parámetros **Z**:

$$
Z_{11}=\frac{V_1}{I_1}|_{I_2=0}=R_1
\qquad
Z_{12}=\frac{V_1}{I_2}|_{I_1=0}=0
$$

$$
Z_{22}=\frac{V_2}{I_2}|_{I_1=0}=0
\qquad
Z_{21}=\frac{V_2}{I_1}|_{I_2=0}=-R_2
$$

Por lo que la **matriz Z** es:

$$
Z=
\begin{bmatrix}
R_1   &   0  \\
0   &   -R_2     
\end{bmatrix}
$$
