#### **Transferencias Bessel**

Para los filtro Bessel tenemos la siguiente expresión utilizando el método de Storch:

$$
T(s)= \frac {K} {senh(s)+cosh(s)}
$$

Dividiendo un termino por el otro:

$$
cotgh(s)=\frac{cosh(s)}{senh(s)}=\frac{1}{s} + \frac{1}{ \frac{3}{s} + \frac{1}{\frac{5}{s}+\frac{1}{...}}}
$$

Para un determinado **$n$** la $cotgh(s)$ se trunca.

Para $n=2$:

$$
cotgh(s)=\frac{1}{s} + \frac{1}{ \frac{3}{s}} = \frac{1}{s} + \frac{s}{3} = \frac{3+s^2}{3s}
$$

$$
T_2(s)= \frac {3}{s^2+3s+3}
$$

Para $n=3$:

$$
cotgh(s)=\frac{1}{s} + \frac{1}{ \frac{3}{s} + \frac{s}{5} } = \frac{1}{s} + \frac{5s}{ 15+s^2 } = \frac{15+s^2+5s^2}{s^3+15s}
$$

$$
T_3(s)= \frac {15}{s^3+6s^2+15s+15}
$$

Para $n=4$

$$
cotgh(s)=\frac{1}{s} + \frac{1} { \frac{3}{s} + \frac{1}{ \frac{5}{s} + \frac{s}{7} }} = \frac{1}{s} + \frac{1} { \frac{3}{s} + \frac{7s}{35+s^2}} = \frac{1}{s} + \frac{35s+s^3}{105+3s^2+7s^2 } = \frac{10s^2+105+35s^2+s^4}{10s^3+105s}
$$

$$
T_4(s)= \frac {105}{s^4+10s^3+45s^2+105s+105}
$$

Para saber cual de ellas cumple con $\alpha_{max}=1dB$ calculamos el módulo de cada una evaluada en $w=1$:

Para orden 2:
$$
|T_2(jw)|=\frac{3}{\sqrt{(3-w^2)^2+(3w)^2}}
$$

$$
|T_2(1)|=0.832
$$

$$
\alpha_{dB}(1)=-20log(0.832)=1.597dB
$$

Como $\alpha_{dB}(1)>1dB$ no cumple con la plantilla 

Para orden 3:
$$
|T_3(jw)|=\frac{15}{\sqrt{(15-6w^2)^2+(15w-w^3)^2}}
$$

$$
|T_3(1)|=0.901
$$

$$
\alpha_{dB}(1)=-20log(0.901)=0.905dB
$$

Como $\alpha_{dB}(1)<1dB$ cumple con la plantilla 

Para orden 4:
$$
|T_4(jw)|=\frac{105}{\sqrt{(w^4-45w^2+105)^2+(105w-10w^3)^2}}
$$

$$
|T_4(1)|=0.93
$$

$$
\alpha_{dB}(1)=-20log(0.93)=0.63dB
$$

Como $\alpha_{dB}(1)<1dB$ cumple con la plantilla

#### **Retardo de Grupo**

Para el *Retardo de Grupo* utilizamos la siguiente expresión:

$$
D(w)=\sum_{i=1}^{n} \frac{-\sigma_i}{\sigma_i^2+(w-w_i)^2}
$$

Donde $\sigma_i$ es la parte real de un polo y $w_i$ es la parte imaginaria de un polo.

En la transferencia $T_3(s)$ tenemos 3 polos:

$P_1=-2.32+j0$

$P_{2,3}=-1.84\pm j1.754$

Por lo que $D_3(w)$ queda de la siguiente manera:

$$
D_3(w)=\frac{2.32}{(-2.32)^2+w^2}+\frac{1.84}{(-1.84)^2+(w-1.754)^2}+\frac{1.84}{(-1.84)^2+(w+1.754)^2}
$$

Para $w=2.5$ y $w=1$:

$$
D_3(2.5)=0.7518
$$

$$
D_3(1) \approx 1
$$

Y el error porcentual queda:

$$
e=\left| \frac{D_3(2.5)-D_3(0)}{D_3(2.5)} \right| *100 = 24.82%
$$

#### **Circuito normalizado con estructuras Sallen-Key**

La función transferencia del circuito Sallen-Key presentado es:

$$
H_1(s)=\frac{ \frac{1}{C1C2R1R2} }{s^2 + s \frac{R1+R2}{C1R1R2} + \frac{1}{C1C2R1R2} }
$$

Como es de segundo orden no alcanza para implementar el filtro Bessel de orden 3, por lo que hay que sumarle un filtro de primer orden:

$$
H_2(s)=\frac{ \frac{1}{RC} }{s + \frac{1}{RC} }
$$

Por lo que la transferencia completa es la siguiente:

$$
T(s)= \frac{ \frac{1}{RC} }{s + \frac{1}{RC} } \frac{ \frac{1}{C1C2R1R2} }{s^2 + s \frac{R1+R2}{C1R1R2} + \frac{1}{C1C2R1R2} }
$$

Suponiendo que $R=R1=R2=1$ nos queda:

$$
T(s)= \frac{ \frac{1}{C} }{s + \frac{1}{C} } \frac{ \frac{1}{C1C2} }{s^2 + s \frac{2}{C1} + \frac{1}{C1C2} }
$$

Para calcular los valores de los capacitores comparamos las secciones de $T(s)$ con $T_3(s)$ empezando con la de primer orden:

$$
\frac{ \frac{1}{C} }{s + \frac{1}{C} }=\frac{ 2.322 }{s + 2.322 }
$$

Donde $\frac{1}{C}=2.322$ por lo que $C=0.4306$

Para la sección de segundo orden:

$$
\frac{ \frac{1}{C1C2} }{s^2 + s \frac{2}{C1} + \frac{1}{C1C2} }=\frac{ 6.459 }{s^2 + s 3.678 + 6.459 }
$$

Donde $\frac{2}{C1}=3.678$ siendo entonces $C1=0.5438$ y, conociendo este valor, obtenemos $C2$ mediante $\frac{1}{C1C2}=6.459$, por lo que $C2=0.2847$

Finalmente obtenemos los valores de los componentes normalizados del circuito:

$R=R1=R2=1$

$C1=0.5438$

$C2=0.2847$

#### **Desnormalización de los componentes**

Nos piden desnormalizar los componentes para lograr un $D(0)=200\micro s$. Para ello tomamos como norma de frecuencia (de retardo en este caso) $\Omega_D=\frac{1}{t_0}$ donde $t_0=200\micro s$ por lo que $\Omega_D=5000 \frac{rad}{seg}$. Como norma de impedancia tomamos $\Omega_Z=1000 \Omega$ 

Teniendo ambas normas procedemos a desnormalizar con las siguientes expresiones:

$$
R=R_n \Omega_Z
$$

$$
C=C_n \frac{1}{\Omega_D \Omega_Z}
$$

Los valores finales entonces son:

$R=R1=R2=1K\Omega$

$C=86.12 nf$

$C1=108.76 nf$

$C2=56.94 nf$
