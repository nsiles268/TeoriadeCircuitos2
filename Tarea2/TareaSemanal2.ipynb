### **Transferencia y párametros**

Para obtener la transferencia nombramos **$V_X$** al nodo entre **$R3$** y la salida de **U1**, y como **$V_Y$** al nodo entre **$C$** y **$R4$**.

Aplicando las ley de corrientes de Kirchhoff nos quedan tres ecuaciones:

$$
\frac{V_i}{R1}=-V_X SC-\frac{V_X}{R2}-\frac{V_o}{R3}\qquad (1)
$$

$$
\frac{V_o}{R4}=-\frac{V_Y}{R4} \qquad (2)
$$

$$
\frac{V_X}{R3}=-V_Y SC\qquad (3)
$$

De **(2)** obtenemos **$V_Y=-V_o$**

De **(3)** obtenemos **$V_X==-V_Y SCR3$**

Reemplazando **(2)** en **(3)** hallamos **$V_X=V_o SCR3$**

Sustituyendo **$V_X$** en **(1)**:

$$
\frac{V_i}{R1}=-V_o S^2C^2R3-V_o\frac{SCR3}{R2}-\frac{V_o}{R3}
$$
$$
\frac{V_i}{R1}=-V_o (S^2C^2R3+\frac{SCR3}{R2}+\frac{1}{R3})
$$
$$
\frac{V_i}{V_o}= -\frac{1}{R1} \frac { \frac{1}{C^2R3} } { S^2+S\frac{1}{CR2}+\frac{1}{C^2R3^2} }
$$

Finalmente la función transferencia es:

$$
T(S)= -\frac{R3}{R1} \frac { \frac{1}{C^2R3^2} } { S^2+S\frac{1}{CR2}+\frac{1}{C^2R3^2} }
$$

En función de  $w_0$ y $Q$:
**$$
T(S)= k \frac { w_0^2 } { S^2+S\frac{w_0}{Q}+w_0^2}
$$**

Siendo:

$k=-\frac{R3}{R1}$

$w_0=\frac{1}{CR3}$

$Q=\frac{R2}{R3}$

### **Circuito pasabanda**

Para obtener un circuito *pasabanda* unicamente debemos cambiar la salida que tomamos. Por lo tanto, ahora ubicaremos **$V_o$** entre **$R3$** y la salida del amplificador operacional **U1**.

Nombrando ahora como **$V_X$** al nodo que antes era **$V_o$** y aplicando las leyes de Kirchhoff:

$$
\frac{V_i}{R1}=-V_o SC-\frac{V_o}{R2}-\frac{V_x}{R3}\qquad (1)
$$

$$
V_Y=-V_X \qquad (2)
$$

$$
\frac{V_o}{R3}=-V_Y SC\qquad (3)
$$

Reemplazando **(2)** en **(3)** hallamos **$V_X= \frac{V_o} {SCR3}$**

Sustituyendo **$V_X$** en **(1)**:
$$
\frac{V_i}{R1}=-V_o SC-\frac{V_o}{R2}-\frac{V_o}{SCR3^2}
$$

$$
\frac{V_i}{R1}=-V_o (SC+\frac{1}{R2}+\frac{1}{SCR3^2})
$$

$$
\frac{V_i}{R1}=-V_o \frac{ S^2C^2R3^2R2 +SCR3^2 +R2 }{SCR2R3^2}
$$

$$
\frac{V_o}{V_i}=-\frac{1}{R1} \frac { S\frac{1}{C} }{ S^2 +S\frac{1}{CR2} + \frac{1}{C^2R3^2} }
$$

Finalmente la función transferencia es:
$$
\frac{V_o}{V_i}=-\frac{R2}{R1} \frac { S\frac{1}{CR2} }{ S^2 +S\frac{1}{CR2} + \frac{1}{C^2R3^2} }
$$

En función de  $w_0$ y $Q$:
**$$
H(S)= k \frac { S\frac{w_0}{Q} } { S^2+S\frac{w_0}{Q}+w_0^2}
$$**

Siendo:

$k=-\frac{R2}{R1}$

$w_0=\frac{1}{CR3}$

$Q=\frac{R2}{R3}$

### **Filtro pasabajos Chebyshev**

Para calcular el $\epsilon $ utilizamos la siguiente expresión:

$$
\epsilon^2 = 10^{ \frac { \alpha_{max} } {10} }-1
$$

Por lo que: 

$\epsilon^2=0.0964$

$\epsilon=0.3106$

Para averiguar el valor de "n" iteramos la siguiente expresión:

$$
\alpha_{db} = 10*log(1 + \epsilon^2 cosh^2( n*cosh^{-1}(w) ) )
$$

Para $n=5$ , $\alpha_{min}=54.3$ por lo tanto:

$n=5$

Para un filtro de aproximación Chebyshev:

$$
|T(jw)|^2 = \frac {1} {1 + \epsilon^2 * C_n ^2 (w)}
$$

donde en nuestro caso:

$C_5(w)=16w^5 - 20w^3 + 5w$

Reemplazando:

$$
|T(jw)|^2 = \frac {1} {1 + \epsilon^2 *(16w^5 - 20w^3 + 5w)^2}
$$

$$
|T(jw)|^2 = \frac {1} {1 + \epsilon^2 * ( 256w^{10} - 640w^8 + 560w^6 - 200w^4 +25w^2) }
$$

$$
|T(jw)|^2 = \frac { \frac {1} {\epsilon^2 * 256} } { w^{10} - 2.5w^8 + \frac{35}{16} w^6 - \frac{25}{32} w^4 + \frac{25}{216} w^2 + \frac {1} {\epsilon^2 * 256} }
$$

Sustituyendo $ w=\frac{s}{j}$
$$
|T(s)|^2 = \frac { -\frac {1} {\epsilon^2 * 256} } { s^{10} + 2.5s^8 + \frac{35}{16} s^6 + \frac{25}{32} s^4 + \frac{25}{216} s^2 - \frac {1} {\epsilon^2 * 256} }
$$

Utilizamos python para obtener las raices y la transferencia de quinto orden:

```python
# Módulos numéricos y de funciones científicas
import numpy as np
from scipy import signal as sig
import sympy as sp
from IPython.display import display, Math

# Módulos para los gráficos
import matplotlib as mpl
from matplotlib import pyplot as plt

# Funciones de PyTC2
from pytc2.sistemas_lineales import bodePlot, pzmap, GroupDelay, analyze_sys
```

```python
#Epsilon
eps=np.sqrt(10**(0.4/10)-1)

#Raices
raices = np.roots([1,0,5/2,0,35/16,0,25/32,0,25/256,0,-1/((eps**2)*256)])
roots = raices[np.real(raices) < 0]

#Coeficientes
coeff = np.poly(roots)
coeff = np.round(coeff,4)

k = np.sqrt(1/((eps**2)*(16**2)))
k = np.round(k,4)

#Variable "S"
s = sp.symbols('s')

#Transferencia de quinto orden
den = sp.Poly(coeff,s)
Ts  = sp.simplify(k) / sp.simplify(den)

display(Math(r'T(s) = ' + sp.latex(Ts)))
```

Para construir el filtro utilizando 2 redes de segundo orden y una de primer orden necesitamos los polos:

```python
#Redondeo a 4 decimales
roots = np.round(roots,4)

#Muestro los polos
for i, r in enumerate(roots):
    print(f"Polo {i+1} = {r.real} + {r.imag}")
```

Con los polos obtenemos los coeficientes de los polinomios

```python
#Coeficientes de las secciones
pol1 = np.array([roots[0], roots[1]])
pol2 = np.array([roots[2], roots[3]])
pol3 = roots[4]

print(np.round(np.poly(pol1),4))
print(np.round(np.poly(pol2),4))
print(np.abs(pol3))
```

Entonces la funcion transferencia queda de la siguiente manera:

$$
T(s)= \frac {1} {s+0.3861} \frac {K_1*1.0536} {s^2 + 0.2386s + 1.0536} \frac {K_2*0.4946} {s^2 + 0.6248s + 0.4946}
$$

Para calcular $K_1$ y $K_2$ igualamos el numerador de esta función trasnferencia con el numerador de la función de quinto orden:

$$
(K_1*1.0536)*(K_2*0.4946)=0.2012
$$

Asumiendo que $K_2 = \frac{K_1}{2}$:

$$
\frac{K_1^2}{2}*0.5211=0.2012
$$

Entonces:

$K_1=0.8787$

$K_2=0.43935$

Finalmente:

$$
T(s)= \frac {1} {s+0.3861}* \frac {0.8787*1.0536} {s^2 + 0.2386s + 1.0536} \frac {0.43935*0.4946} {s^2 + 0.6248s + 0.4946}
$$

Para corroborar este diseño graficamos el diagrama de bode:

```python
#Secciones
H1=sig.TransferFunction(0.8787*1.0536,np.poly(pol1))
H2=sig.TransferFunction(0.43935*0.4946,np.poly(pol2))
H3=sig.TransferFunction(1,[1,np.abs(pol3)])

#Numerador y denominador total
num = np.polymul(np.polymul(H1.num, H2.num), H3.num)
den = np.polymul(np.polymul(H1.den, H2.den), H3.den)

#Función trasnferencia
TS=sig.TransferFunction(num,den)

#Diagrama de bode
_, axes = bodePlot(TS)
axes[0].set_ylim(-0.5,0.2)
plt.show()
```

Observamos que módulo oscila entre $0$ y $-0.4 dB$ por lo que cumple con lo pedido en la banda de paso

```python
#Diagrama de bode
_, axes = bodePlot(TS)
axes[0].set_xlim(1,3)
plt.show()
```

Vemos también que para $w=3$ hubo una caida mayor a $48 dB$ por lo que se cumple el $\alpha_{min}$ para el inicio de la banda de stop.

### **Red pasabanda normalizada y simulaciones**

Siendo la función transferencia:

$$
H(S)=-\frac{R2}{R1} \frac { S\frac{1}{CR2} }{ S^2 +S\frac{1}{CR2} + \frac{1}{C^2R3^2} }
$$

Procedemos a normalizar con $\Omega_w= \frac{1}{CR3}$ por lo que la transferencia queda:

$$
H($)=-\frac{R2}{R1} \frac { $\frac{R3}{R2} }{ $^2 +$\frac{R3}{R2} + 1 }
$$

Y si $R1=R2=R3=1$:

$$
H($)=-1 \frac { $1 }{ $^2 +$1 + 1 }
$$

Siendo:

$Q=\frac{R2}{R3}=1$ 

$k=-\frac{R2}{R1}=-1$ 

$w_0=\frac{1}{CR3}=1 $ 

```python
#Componentes normalizados
#R1=1
#R2=1
#R3=1
#k=-R2/R1

#Transferencia con componentes
#num=np.array([k* R3/R2 ,0])
#den=np.array([1, R3/R2 ,1])

#Transferencia normalizada
num=np.array([1,0])
den=np.array([1,1,1])

H= sig.TransferFunction(num,den)

bodePlot(H)
plt.show()
```

Tambien podria quedar en función de $Q$, ya que $\frac{R3}{R2}=\frac{1}{Q}$:

$$
H($)=-\frac{R2}{R1} \frac { $\frac{1}{Q} }{ $^2 +$\frac{1}{Q} + 1 }
$$

Y asumiendo que $R1=R3=1$ nos queda que $Q=R2$ por lo que la funcion transferencia queda:

$$
H($)=-Q \frac { $\frac{1}{Q} }{ $^2 +$\frac{1}{Q} + 1 }
$$

```python
Q=1
k=-Q

num=np.array([ k* 1/Q , 0])
den=np.array([1 , 1/Q , 1])

H= sig.TransferFunction(num,den)

bodePlot(H)
plt.show()
```
